```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv('datos.csv', delimiter=";")
```

```python
df.head()
```

```python
dbscan = DBSCAN()
dbscan.fit(df)
labels = dbscan.labels_
```

```python
plt.scatter(df['Variable1'], df['Variable2'], c=labels, cmap='plasma')
plt.show()
```
