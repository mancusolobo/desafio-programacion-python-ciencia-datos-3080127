```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('clusters.csv', delimiter=";")
```

```python
df.head()
```

```python
clusters = KMeans(n_clusters=4, n_init='auto')
clusters.fit(df)
y_clusters = clusters.predict(df)
cluster_centers = clusters.cluster_centers_
```

```python
plt.scatter(df['Latitud'], df['Longitud'], c=y_clusters, s=20)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200)
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.show()
```
