```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import normalize
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv('clientes.csv', delimiter=";")
```

```python
df.head()
```

```python
dendrogram(linkage(df, method='ward', metric='euclidean'))
plt.axhline(y=15000, color='r', linestyle='--')
plt.show()
```

```python
clusters = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
clusters.fit(df)
labels = clusters.labels_
```

```python
plt.scatter(df['Salario'], df['Gasto'], c=labels)
plt.xlabel('Salario ($)')
plt.ylabel('Gasto (%)')
plt.show()
```
